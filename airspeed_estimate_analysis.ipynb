```python
from px4tools import logdynamicpressure as ldp
import numpy as np
from bokeh.plotting import figure, output_notebook, show
output_notebook()
```

```python
file = '/home/dennis/src/px4Analysis/log/altitude/12_52_01.ulg'
```

```python
# load data and do airspeed estimation
df = ldp.load_data(file)
df = ldp.estimate_airspeed(df)
```

```python
p_global = figure(title='airspeed global',plot_width = 800, plot_height = 500 )
p_body = figure(title='airspeed body',plot_width = 800, plot_height = 500 )
p_mag = figure(title='airspeed magnitude',plot_width = 800, plot_height = 500 )
```

```python
p_global.line(df.index, df.t_vehicle_local_position_0__f_vx, color = 'red', legend='v_x',line_width = 2)
p_global.line(df.index, df.t_vehicle_altitude_hold_0__f_v_airx, color = 'black', legend='v_air_x',line_width = 2)
p_global.line(df.index, df.t_vehicle_local_position_0__f_vy, color = 'orange', legend='v_y',line_width = 2)
p_global.line(df.index, df.t_vehicle_altitude_hold_0__f_v_airy, color = 'blue', legend='v_air_y',line_width = 2)
```

```python
show(p_global, notebook_handle=True)
```

```python
p_body.line(df.index, df.t_vehicle_local_position_0__f_vx_body, color = 'red', legend='v_x_body',line_width = 2)
p_body.line(df.index, df.t_vehicle_altitude_hold_0__f_v_airx_body, color = 'black', legend='v_air_x_body',line_width = 2)
p_body.line(df.index, df.t_vehicle_local_position_0__f_vy_body, color = 'orange', legend='v_y_body',line_width = 2)
p_body.line(df.index, df.t_vehicle_altitude_hold_0__f_v_airy_body, color = 'blue', legend='v_air_y_body',line_width = 2)
```

```python
show(p_body, notebook_handle=True)
```

```python
p_mag.line(df.index, np.sqrt(df.t_vehicle_local_position_0__f_vx**2 + 
                             df.t_vehicle_local_position_0__f_vy**2), color = 'red', legend='v_mag',line_width = 2)
p_mag.line(df.index, np.sqrt(df.t_vehicle_altitude_hold_0__f_v_airx**2 +
                             df.t_vehicle_altitude_hold_0__f_v_airy**2), color = 'black', legend='v_air_mag',line_width = 2)
```

```python
show(p_mag, notebook_handle=True)
```
